```python
# keep only admin_code and country columns
area_id_country = df[['admin_code', 'country']].drop_duplicates().reset_index(drop=True)
```

```python
# rename admin_code to area_id
area_id_country = area_id_country.rename(columns={'admin_code': 'area_id'})
```

```python
country_regions = {
    # --- West Africa ---
    'Benin': 'West Africa',
    'Burkina Faso': 'West Africa',
    'Cabo Verde': 'West Africa',
    "Cote d'Ivoire": 'West Africa',
    'Gambia': 'West Africa',
    'Ghana': 'West Africa',
    'Guinea': 'West Africa',
    'Guinea-Bissau': 'West Africa',
    'Liberia': 'West Africa',
    'Mali': 'West Africa',
    'Mauritania': 'West Africa',
    'Niger': 'West Africa',
    'Nigeria': 'West Africa',
    'Senegal': 'West Africa',
    'Sierra Leone': 'West Africa',
    'Togo': 'West Africa',

    # --- Central Africa ---
    'Cameroon': 'Central Africa',
    'Central African Republic': 'Central Africa',
    'Chad': 'Central Africa',
    'Democratic Republic of the Congo': 'Central Africa',

    # --- East Africa ---
    'Burundi': 'East Africa',
    'Djibouti': 'East Africa',
    'Ethiopia': 'East Africa',
    'Kenya': 'East Africa',
    'Somalia': 'East Africa',
    'South Sudan': 'East Africa',
    'Sudan': 'East Africa',
    'Tanzania': 'East Africa',
    'Uganda': 'East Africa',

    # --- South Africa (Region) ---
    # Note: Includes SADC nations often grouped here for economic analysis
    'Angola': 'South Africa',
    'Eswatini': 'South Africa',
    'Lesotho': 'South Africa',
    'Madagascar': 'South Africa',
    'Malawi': 'South Africa',
    'Mozambique': 'South Africa',
    'Namibia': 'South Africa',
    'South Africa': 'South Africa',
    'Zambia': 'South Africa',
    'Zimbabwe': 'South Africa',

    # --- Latin America ---
    'Dominican Republic': 'Latin America',
    'Ecuador': 'Latin America',
    'El Salvador': 'Latin America',
    'Guatemala': 'Latin America',
    'Haiti': 'Latin America',
    'Honduras': 'Latin America',

    # --- Asia ---
    'Afghanistan': 'Asia',
    'Bangladesh': 'Asia',
    'East Timor': 'Asia',
    'Lebanon': 'Asia',
    'Pakistan': 'Asia',
    'Palestinian Territory': 'Asia',
    'Tajikistan': 'Asia',
    'Yemen': 'Asia'
}
# map country to region
area_id_country['region'] = area_id_country['country'].map(country_regions)
```

```python
# drop country
area_id_country = area_id_country.drop(columns=['country'])

# encode region as integer
region_mapping = {
    'Asia': 0,
    'East Africa': 1,
    'West Africa': 2,
    'South Africa': 3,
    'Central Africa': 4,
    'Latin America': 5
}

area_id_country['region'] = area_id_country['region'].map(region_mapping)
```

```python
# save mapping
area_id_country.to_csv('area_id_country_region_mapping.csv', index=False)
```
